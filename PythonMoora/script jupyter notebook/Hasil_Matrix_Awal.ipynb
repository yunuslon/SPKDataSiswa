```python
import pandas as pd
import numpy as np
import math
from pandas import DataFrame, read_csv
```

```python
Hasil_Tes_MTK= pd.read_csv('Hasil_Tes_MTK.csv')
cols=['Hasil Tes MTK']
hasilHTmtk=Hasil_Tes_MTK[cols]
hasilHTmtk

def hasil_tes_Mtk(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilHTmtk)):
        a=(math.pow(hasilHTmtk['Hasil Tes MTK'][y],2))
        b = b+a
    for i in range(len(hasilHTmtk)):
        s = hasilHTmtk['Hasil Tes MTK'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
    return tampung
a=hasil_tes_Mtk()
hasil=pd.DataFrame(data=a,columns=['Nilai'])
hasil
```

```python
Hasil_MTK= pd.read_csv('MTK.csv')
cols=['MTK']
hasilM=Hasil_MTK[cols]
hasilM

def hasil_mtk(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilM)):
        a=(math.pow(hasilM['MTK'][y],2))
        b = b+a
    for i in range(len(hasilM)):
        s = hasilM['MTK'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
    return tampung
a=hasil_mtk()
hasil=pd.DataFrame(data=a,columns=['Nilai'])
hasil
```

```python
Hasil_Karakter= pd.read_csv('Karakter.csv')
cols=['Karakter']
hasilK=Hasil_Karakter[cols]
hasilK

def hasil_karakter(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilK)):
        a=(math.pow(hasilK['Karakter'][y],2))
        b = b+a
    for i in range(len(hasilM)):
        s = hasilK['Karakter'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
    return tampung
a=hasil_karakter()
hasil=pd.DataFrame(data=a,columns=['Nilai'])
hasil
```

```python
Hasil_Pengalaman_Lomba= pd.read_csv('Pengalaman_lomba.csv')
cols=['Pengalaman Lomba']
hasilPL=Hasil_Pengalaman_Lomba[cols]
hasilPL

def hasil_plomba(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilPL)):
        a=(math.pow(hasilPL['Pengalaman Lomba'][y],2))
        b = b+a
    for i in range(len(hasilPL)):
        s = hasilPL['Pengalaman Lomba'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
    return tampung
a=hasil_plomba()
hasil=pd.DataFrame(data=a,columns=['Nilai'])
hasil
```

```python
Hasil_Kelas= pd.read_csv('Kelas.csv')
cols=['kelas']
hasilKls=Hasil_Kelas[cols]
hasilKls

def hasil_kelas(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilKls)):
        a=(math.pow(hasilKls['kelas'][y],2))
        b = b+a
    for i in range(len(hasilKls)):
        s = hasilKls['kelas'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
    return tampung
a=hasil_kelas()
hasil=pd.DataFrame(data=a,columns=['Nilai'])
hasil
```

```python
def Hasil_Akhir():
    Mtk=hasil_mtk()
    Kelas=hasil_kelas()
    Karakter=hasil_karakter()
    PLomba=hasil_plomba()
    TesMtk=hasil_tes_Mtk()
    Matrix=list(zip(Mtk,Karakter,PLomba,TesMtk,Kelas))
    df = pd.DataFrame(data = Matrix, columns=['Matematika', 'Karakter','Pengalaman Lomba','Hasil Tes MTK','Kelas'])
    Hasil_Nama= pd.read_csv('Kelas.csv')
    Nama=Hasil_Nama['nama']
    new=pd.concat([Nama,df],axis=1)
    return new
```

```python
HA=Hasil_Akhir()
```

```python
HA.to_csv('Matrix_Awal.csv',index=False)
```
