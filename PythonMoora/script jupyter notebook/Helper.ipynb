```python
import pandas as pd
import numpy as np
import math
from pandas import DataFrame, read_csv
```

```python
Hasil_Tes_MTK= pd.read_csv('Hasil_Tes_MTK.csv')
cols=['Hasil Tes MTK']
hasilHTmtk=Hasil_Tes_MTK[cols]
hasilHTmtk

def Hasil_Tes_Mtk(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilHTmtk)):
        a=(math.pow(hasilHTmtk['Hasil Tes MTK'][y],2))
        b = b+a
    for i in range(len(hasilHTmtk)):
        s = hasilHTmtk['Hasil Tes MTK'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
        HTM=pd.DataFrame(data=tampung,columns=['Nilai'])
    return HTM



#________________________________________________#

Hasil_MTK= pd.read_csv('MTK.csv')
cols=['MTK']
hasilM=Hasil_MTK[cols]
hasilM

def Hasil_Mtk(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilM)):
        a=(math.pow(hasilM['MTK'][y],2))
        b = b+a
    for i in range(len(hasilM)):
        s = hasilM['MTK'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
        MTK=pd.DataFrame(data=tampung,columns=['Nilai'])
        
    return MTK


#________________________________________________#

Hasil_Karakter= pd.read_csv('Karakter.csv')
cols=['Karakter']
hasilK=Hasil_Karakter[cols]
hasilK

def Hasil_Karakter(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilK)):
        a=(math.pow(hasilK['Karakter'][y],2))
        b = b+a
    for i in range(len(hasilM)):
        s = hasilK['Karakter'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
        Karakter=pd.DataFrame(data=tampung,columns=['Nilai'])
    return Karakter


#________________________________________________#

Hasil_Pengalaman_Lomba= pd.read_csv('Pengalaman_lomba.csv')
cols=['Pengalaman Lomba']
hasilPL=Hasil_Pengalaman_Lomba[cols]
hasilPL

def Hasil_Plomba(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilPL)):
        a=(math.pow(hasilPL['Pengalaman Lomba'][y],2))
        b = b+a
    for i in range(len(hasilPL)):
        s = hasilPL['Pengalaman Lomba'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
        Plomba=pd.DataFrame(data=tampung,columns=['Nilai'])
        
    return Plomba

#________________________________________________#

Hasil_Kelas= pd.read_csv('Kelas.csv')
cols=['kelas']
hasilKls=Hasil_Kelas[cols]
hasilKls

def Hasil_Kelas(y=0):
    b = 0
    tampung=[]
    for y in range(len(hasilKls)):
        a=(math.pow(hasilKls['kelas'][y],2))
        b = b+a
    for i in range(len(hasilKls)):
        s = hasilKls['kelas'][i]
        ad=s/(math.sqrt(b))
        tampung.append(ad)
        Kelas=pd.DataFrame(data=tampung,columns=['Nilai'])
    return Kelas

Nama= pd.read_csv('Kelas.csv')
def siswa ():
    hasilnama=Nama['nama']
    return hasilnama

def Hasil_Matrix_Awal():
    Mtk=Hasil_Mtk()
    Kelas=Hasil_Kelas()
    Karakter=Hasil_Karakter()
    PLomba=Hasil_Plomba()
    TesMtk=Hasil_Tes_Mtk()
    Nama=siswa()
    Matrix=pd.concat([Nama,Mtk,Karakter,Plomba,TesMtk,Kelas],axis=1)
    Matrix.columns=['Nama','Matematika', 'Karakter','Pengalaman Lomba','Hasil Tes MTK','Kelas']

    return Matrix

```
